```python
import pandas as pd #数据分析
import numpy as np #科学计算
#用现有数据构造一个dataframe
data=pd.read_csv("/Users/xwx/Desktop/kaggle_titanic/data/train.csv")
data2=pd.read_csv("/Users/xwx/Desktop/kaggle_titanic/data/test.csv")
data=data[0:15]
data2=data[0:15]
```

```python
#select Age from data where Sex="male" AND Pclass=3

data['Age'][(data['Sex']=='male') & (data.Pclass==3)]
#上式子与下式等价
#data['Age'][data['Sex']=='male'] [data.Pclass==3]
```

```python
#select * from data where Cabin IS NULL
data[data.Cabin.isna()]

#select * from data where Cabin IS NOT NULL
data[data.Cabin.notna()]
```

```python
#select * from data where Pclass in (1,2)
data[data.Pclass.isin([1,2])]
```

```python
#select Survived,Age from data group by Sex order by Age
p=data.groupby([data.Sex,data.Pclass])#["Survived","Age"]#groupby完之后数据类型变为DataFrameGroupBy对象，可以接sum（），mean（）等
#p=p.apply(lambda x:x.sort_values(by='Age',ascending=False))#order by Age 其中ascending代表升序或降序，这时p的数据结构变回dataframe
#print(type(p)) #查看p的数据结构
df=pd.DataFrame(p.count()['PassengerId'])
df.head()
#关于groupby的一些tips，groupby非常特殊，处理之后会更改数据结构变成DataFrameGroupBy，因此很多属性都不能用了，所以写的时候最好把它再调换回DataFrame
```

```python
#select PassengerId,(case when Pclass=1 then '高等仓' else '低等舱') as 等级 from data
data['等级']=data.Pclass.apply(lambda x:'高等仓' if x==1 else "低等舱")
data
```

```python
#select A.*,B.* from table A left join table2 B on A.colA = B.colA
#select data.*.data2.* from data left join data2 on data.Pclass=data2.Pclass
pd.merge(data,data2,how='left',on='Pclass')
#merge更多用法见 https://blog.csdn.net/LY_ysys629/article/details/73849543
```

```python
#data union all data2
pd.concat([data,data2])
#concat更多用法见https://blog.csdn.net/zhongjunlang/article/details/79604499
```

```python
#select Pclass,Name,(Age+Parch) as AgeP from data 
data['AgeP']=data.apply(lambda x:x.Age+x.Parch,axis=1)
#tips:1.新建的列，不可用data.AgeP 必须用data['AgeP'] 2.apply中axis代表函数应用于每行，默认axis=0，代表应用于每列
data.head(10)
```
