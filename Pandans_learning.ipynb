```python
data=pd.read_csv("/Users/xwx/Desktop/kaggle_titanic/data/train.csv")
#data=pd.read_csv("/Users/xwx/Desktop/kaggle_titanic/data/test.csv")
data.info()
```

```python
#创造数据
df1=data[0:15][['Survived','Sex','Name']]
df2=data[15:30][['Pclass','PassengerId','Parch']]
df2.head()
```

```python
#重命名

#只对其中某列重命名
df1.rename(columns={"Survived":'获救？'})

#对所有列重命名
df1.columns=['是否获救','性别','姓名']
df1.head()
```

```python
#连接俩dataframe

#axis=0代表行连接
pd.concat([df1,df2],axis=0) #按照列名相同的合并 shape=15*6

#axis=1代表列链接
pd.concat([df1,df2],axis=1)#按照index相同连接 shape=30*6

#将df2的索引列重新从0开始排列（drop=1代表不保存原来的index）再次连接试试
pd.concat([df1,df2.reset_index(drop=1)],axis=1)#shape=15*6


#合并的同时区分哪些数据来自哪里
pd.concat([df1,df2],axis=0,keys=['df1','df2'])

```

# 小知识

```python
#hasattr()判断对象是否有对应的属性
#创建一个类
class Coordinate:
    x = 10
    y = -5
    z = 0
#创建一个实例
point1 = Coordinate() 
print(hasattr(point1, 'x'))
print(hasattr(point1, 'y'))
print(hasattr(point1, 'z'))
print(hasattr(point1, 'no'))  # 没有该属性
```
